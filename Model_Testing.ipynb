```python
import cv2
import tensorflow as tf
from keras.preprocessing import image
from PIL import Image
import numpy as np
```

```python
model = tf.keras.models.load_model(r'ImageClassification_model')
```

```python
def names(number):
    if number == 0:
        return 'Its a buildings'
    elif number == 1:
        return 'Its a forest'
    elif number == 2:
        return 'Its a glacier'
    elif number == 3:
        return 'Its a mountain'
    elif number == 4:
        return 'Its a sea'
    else:
        return 'Its a street'
```

```python
# model = tf.keras.models.load_model(r'BrainTumor_model')

from matplotlib.pyplot import imshow

img = Image.open(r"C:\Users\meabh\Deep Learning\Image_Classification\498.jpg")

x = np.array(img.resize((150,150)))
x = x.reshape(1,150,150,3)
res = model.predict_on_batch(x)
classification = np.where(res == np.amax(res))[1][0]
imshow(img)
print(str(names(classification)))
```
